```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the dataset
file_path = 'path_to_csv_file'
df = pd.read_csv(file_path)
```

```python
# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Sort the dataframe by date
df = df.sort_values('Date')
```

```python
# Calculate the trend lines
df['MA_50'] = df['Close'].rolling(window=50).mean()
df['MA_200'] = df['Close'].rolling(window=200).mean()
```

```python
# Plot Closing Price
plt.figure(figsize=(12, 4))
plt.plot(df['Date'], df['Close'], label='Closing Price', color='black')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('HDFCBANK Closing Price')
plt.legend()
plt.show()
```

```python
# Plot 50-day Moving Average
plt.figure(figsize=(12, 4))
plt.plot(df['Date'], df['MA_50'], label='50-day MA', color='blue')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('HDFCBANK 50-day Moving Average')
plt.legend()
plt.show()
```

```python
# Plot 200-day Moving Average
plt.figure(figsize=(12, 4))
plt.plot(df['Date'], df['MA_200'], label='200-day MA', color='red')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('HDFCBANK 200-day Moving Average')
plt.legend()
plt.show()
```

```python
# Plot all together
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Close'], label='Closing Price', color='black')
plt.plot(df['Date'], df['MA_50'], label='50-day MA', color='blue')
plt.plot(df['Date'], df['MA_200'], label='200-day MA', color='red')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('HDFCBANK Stock Price with Trend Lines')
plt.legend()
plt.show()

```

```python
# Highlighting upward and downward trends
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Close'], label='Closing Price', color='black')
plt.plot(df['Date'], df['MA_50'], label='50-day MA', color='blue')
plt.plot(df['Date'], df['MA_200'], label='200-day MA', color='red')

plt.fill_between(df['Date'], df['MA_50'], df['MA_200'], where=(df['MA_50'] > df['MA_200']), interpolate=True, color='green', alpha=0.2)
plt.fill_between(df['Date'], df['MA_50'], df['MA_200'], where=(df['MA_50'] < df['MA_200']), interpolate=True, color='red', alpha=0.2)

plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.title('HDFCBANK Stock Price with Trend Lines')
plt.legend()
plt.show()
```
